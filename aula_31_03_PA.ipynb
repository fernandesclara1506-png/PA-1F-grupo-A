Escrever um programa que recebe o nome e a idade
do usuário e verifica se o usuário é criança, adolecente e adulto

<=12 anos é criança
=< 17 anos é adolecente
Adulto







```python
#entrada
nome = str(input("Digite seu nome: "))
idade = int(input("Digite sua idade: "))

#processamento/ saída
if idade <= 12:
  print(f"Oii {nome}, você tem {idade} anos, então você é criança")

elif idade <= 17:
   print(f"Oii {nome}, você tem {idade} anos, então você é adolecente")

else:
 print(f"Oii {nome}, você tem {idade} anos, então você é adulto")
```

```python
#entrada
energia = int(input("Digite a energia do personagem: "))

#processamento/ saída
if energia <= 30:
  print("Seu personagem está fraco")

elif energia <= 70:
  print("Seu personagem está cansado")

else:
 print("Seu personagem esta forte")
```
