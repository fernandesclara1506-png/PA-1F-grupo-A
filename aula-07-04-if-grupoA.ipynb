Crie um programa que leia a idade de uma pessoa e classifique conforme as regras abaixo:

• Até 17 anos: Menor de idade

• De 18 a 59 anos: Adulto

• 60 anos ou mais: Idoso



```python
#entrada
nome = str(input("Digite seu nome: "))
idade = int(input("Digite sua idade: "))

#processamento/ saída
if idade <= 17:
  print(f"Oii {nome}, você tem {idade} anos, então você é menor de idade")

elif idade <= 59:
   print(f"Oii {nome}, você tem {idade} anos, então você é adulto")

else:
 print(f"Oii {nome}, você tem {idade} anos, então você é idoso")
```

Crie um programa que leia o valor de uma compra e aplique o desconto conforme as regras:

• Até R$100: sem desconto

• De R$101 a R$500: 10% de desconto

• Acima de R$500: 20% de desconto

```python
#entrada
valorcompra = float(input("Digite o valor da compra: R$ "))

desconto = 0

# regras de desconto
if valorcompra <= 100:
   print("Sua compra foi menor que 100, então não terá desconto!")

elif valorcompra <= 500:
    desconto = 0.10

else:
    desconto = 0.20

# Cálculos
valordesconto = valorcompra * desconto
valorfinal = valorcompra - valordesconto

# Exibição dos resultados
print(f"O valor do produto: R$ {valorcompra:}")
print(f"Desconto aplicado nele: {desconto*100:}% (R$ {valordesconto:})")
print(f"Valor total a pagar: R$ {valorfinal:}")
```

Crie um programa que leia a energia de um personagem (de 0 a 100) e classifique:

• 31 a 70: Normal

```python
#entrada
energia = int(input("Digite a energia do personagem: "))

#processamento/ saída
if energia == 0:
  print("Seu personagem está morto")

elif energia <= 30:
    print("Seu personagem está em estado crítico")

elif energia <= 70:
  print("Seu personagem está normal")

else:
 print("Seu personagem esta saudável")
```

Crie um programa que leia o salário de um funcionário e sua pontuação de desempenho. O
programa deve calcular um bônus de acordo com as regras:

• Pontuação menor que 5: sem bônus

• Pontuação de 5 a 7: bônus de 5% do salário

• Pontuação de 8 a 9: bônus de 10% do salário

• Pontuação igual a 10: bônus de 15% do salário

Ao final, mostre o valor do bônus e o salário final com o bônus aplicado.Média menor que 5:
Reprovado

```python
#entrada
salario = float(input("Digite o salário atual do funcionário: R$ "))
pontuacao = int(input("Digite a pontuação de desempenho (0 a 10): "))

bonuseporcentagem = 0

#processamneto/saída (pontuação)
if pontuacao < 5:
    print("Sem bônus,ou melhor, reprovado!")
elif 5 <= pontuacao <= 7:
    bonuseporcentagem = 0.05
elif 8 <= pontuacao <= 9:
    bonuseporcentagem = 0.10
elif pontuacao == 10:
   bonuseporcentagem = 0.15
else:
    print("Pontuação inválida. Digite um valor entre 0 e 10.")

# Calculos
valorbonus = salario * bonuseporcentagem
salariofinal = salario + valorbonus

# saída
print(f"Pontuação: {pontuacao}")
print(f"Salário Base: R$ {salario:}")
print(f"Bônus: {bonuseporcentagem * 100:}% (R$ +{valorbonus:})")
print(f"Salário Final: R$ {salariofinal:}")

```
