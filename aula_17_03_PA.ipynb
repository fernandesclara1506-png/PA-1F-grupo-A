TUDO JUNTO

```python
#entrada
a= int(input("digite um número: "))
b= int(input("digite um número: "))

#Processamento
soma = a + b
subtração = a - b
multiplicação = a * b
divisão = a / b
div_inteira = a // b
resto = a % b
potencia = a ** b

#Saída
print("")
print("-------------------------SOMA-----------------------------")
print(f"o resultado da SOMA de {a} e {b} é: {soma}")
print("")
print("-------------------------SUBTRAÇÃO-----------------------------")
print(f"o resultado da SUBTRAÇÃO de {a} e {b} é: {subtração}")
print("")
print("----------------------MULTIPLICAÇÃO----------------------------")
print(f"o resultado da MULTIPLICAÇÃO de {a} e {b} é: {multiplicação}")
print("")
print("--------------------------DIVISÃO------------------------------")
print(f"o resultado da DIVISÃO de {a} e {b} é: {divisão}")
print("")
print("----------------------DIVISÃO INTEIRA--------------------------")
print(f"o resultado da DIVISÃO INTEIRA de {a} e {b} é: {div_inteira}")
print("")
print("---------------------------RESTO-------------------------------")
print(f"o RESTO da divisão de {a} por {b} é: {resto}")
print("")
print("-------------------------POTÊNCIA------------------------------")
print(f"o resultado da POTENCIA de {a} elevado a {b} é: {potencia}")
```

MAIOR OU MENOR


```python
#entrada
a= int(input("digite um número: "))
b= int(input("digite um número: "))

#Processamento

#comparação
maior = a > b
menor = a < b
maiorigual = a >= b
menorigual = a <= b

# saída
print(maior)
print(menor)
print(maiorigual)
print(menorigual)
```
