1.	Par ou ímpar

Peça um número ao usuário e informe se ele é par ou ímpar.


```python
#entrada
numero= int(input("digite um número: "))
#processamento/saída
if numero % 2 == 0:
  print(f"o número {numero} é par")
else:
    print(f"o número {numero} é ímpar")
```

2.	Maior de dois números

Solicite dois números e mostre qual é o maior (ou se são iguais).


```python
#entrada
numero1 = float(input("digite o primeiro número: "))
numero2 = float(input("digite o segundo número: "))

#processamento
if numero1 > numero2:
  print(f"o primeiro número ({numero1}) é maior que o segundo número ({numero2})")
elif numero1 < numero2:
    print(f"o primeiro número ({numero1}) é menor que o segundo número ({numero2})")
else:
  print(f" os dois números ({numero1} e {numero2}) são iguais ")
```

3.	Aprovação de aluno

Peça a nota de um aluno e informe:

o	Aprovado (nota ≥ 7)

o	Reprovado (nota < 5)



```python
# Entrada
media = float(input("Digite a média final do aluno entre 0 e 10: "))

# Processamento/saida
if media >= 7:
  print("O aluno esta APROVADO!")

elif 5.0 <= media < 6.9:
  print("O aluno está de RECUPERAÇÃO!")

else:
 print("O aluno está REPROVADO!")
```

4.	Classificação de idade

Receba a idade e classifique:

o	Criança (0–12)

o	Adolescente (13–17)

o	Adulto (18–59)

o	Idoso (60+)


```python
#entrada
idade=int(input("digite a sua idade: "))

#procesamento/saída
if idade<=12:
  print("você é uma criança")
elif 13 <= idade <= 17:
  print("você é um adolecente")
elif 18<= idade <= 59:
  print("você é um adulto")
else:
  print("você é um idoso")
```

5.	Senha correta

Defina uma senha fixa no código.
Peça ao usuário para digitar e informe se está correta ou incorreta.


```python
SENHA_CORRETA = 150611

# Entrada
senha = input("Digite a sua senha: ")

# Processamento/Saída
if senha == SENHA_CORRETA:
  print("Senha correta! Acesso concedido.")
elif senha!= SENHA_CORRETA:
    print("Senha incorreta! Acesso negado, tente outra vez.")
else:
  print("Senha incorreta! Acesso negado,tente outra vez.")
```


7.	Calculadora simples

Peça dois números e uma operação (+, -, *, /).

Use if/elif para calcular o resultado.


```python
#entrada
a= int(input("digite um número: "))
b= int(input("digite um número: "))
operacao= input("Digite a operação (+, -, *, /): ")
#processamento
if operacao == "+":
    resultado = a + b
elif operacao == "-":
    resultado = a - b
elif operacao == "*":
    resultado = a * b
elif operacao == "/":
    if b != 0:
        resultado = a / b
    else:
        resultado = "não existe divisão por zero!"
else:
    resultado = "Operação inválida!"

#saída
print("Resultado:", resultado)
```

8.	Dia da semana

Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo).


```python
# Pedindo um número
dia = int(input("Digite um número de 1 a 7: "))

# processamentp
if dia == 1:
    print("É Domingo")
elif dia == 2:
    print("É Segunda-feira")
elif dia == 3:
    print("É Terça-feira")
elif dia == 4:
    print("É Quarta-feira")
elif dia == 5:
    print("É Quinta-feira")
elif dia == 6:
    print("É Sexta-feira")
elif dia == 7:
    print("É Sábado")
else:
    print("Digite de 1 a 7.")
```
