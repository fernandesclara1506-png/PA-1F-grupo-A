Calcular a área do retângulo

```python
# Entrada
comprimento = float(input("digite o comprimento do retângulo: "))
altura = float(input("digite o altura do retângulo: "))

# Processamento
área = comprimento * altura

# Saída
print(f"o total da área do retângulo é: {área}")
```

Programa que calcula o dobro de um número

```python
# Entrada
numero = float(input("digite um número: "))

# Processamento
total = numero * 2

# Saída
print(f"o dobro de {numero} é: {total} ")
```

Escrever um programa que calcula as médias das notas do aluno

```python
# Entrada
nome = str(input("Qual é o nome do aluno? "))
idade = int(input("Qual é a idade do aluno? "))
primeira = float(input("Qual foi a primeira nota do bimestre: "))
segunda = float(input("Qual foi a segunda nota do bimestre: "))
terceira = float(input("Qual foi a terceira nota do bimestre: "))
quarta = float(input("Qual foi a quarta nota do bimestre: "))

# Processamento
total= (primeira + segunda + terceira + quarta)/4

# Saída
print(f"A média do aluno {nome}, com {idade} anos é: {total} ")
```
